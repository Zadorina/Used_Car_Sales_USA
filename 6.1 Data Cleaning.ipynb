```python
# import libraries
import pandas as pd
import numpy as np
import os
import datetime as dt
```

```python
# create path
path = r'C:\Users\Anna\Desktop\Data Analytics\Achievment 6\Raw Data'
```

```python
# import Cars Second Hand Market USA.csv file
df_data = pd.read_csv(os.path.join(path,'Cars Second Hand Market USA.CSV'))
df_data.head()
```

```python
# check the info of dataframe
df_data.info()
```

```python
# check the shape of the Dataframe
df_data.shape
```

```python
# describe the data
df_data.describe()
```

```python
# sorting data by year
df_data.sort_values (by = ['year'], inplace = True, ascending = True)
df_data.head()
```

```python
# checking missing data
df_data.isnull().sum()
```

```python
# changing the column State or Province to State
df_data.rename(columns = { 'State or Province' : 'State'})
```

```python
# check for duplicates
df_data_dups = df_data[df_data.duplicated()]
df_data_dups
```

```python
# check for mixed types
for col in df_data.columns.tolist():
  weird = (df_data[[col]].applymap(type) != df_data[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_data[weird]) > 0:
    print (col)
```

```python
# check for Sold and Not Sold Cars
df_data.sold.value_counts()
```

```python
# removing Not Sold Cars
df=df_data[df_data.sold!="N"]
df.sold.value_counts()
```

```python
df.to_csv(os.path.join(path, 'Used Cars Cleaned.csv'))
```
