```python
# changing the column State or Province to State
df=df_sales.rename(columns = { 'State or Province' : 'State'})
df
```

```python
# Create a scatterplot using matplotlib for another look at how the chosen variables plot against each other.

df.plot(x = 'engine', y='selling_price',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Engine of car vs Selling Price')  
plt.xlabel('engine')  
plt.ylabel('selling_price')  
plt.show()
```

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = df['engine'].values.reshape(-1,1)
y = df['selling_price'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### Hypothesis

The more powerfull the motor of the car is, the higher its cost

### 04. Regression analysis

```python
# Create a regression object.

regression = LinearRegression() 
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
y_predicted = regression.predict(X_test)
```

```python
plot_test = plt
plot_test.scatter(X_test, Y_test, color='lightblue', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Engine of car vs Selling Price (Test set)')
plot_test.xlabel('engine')
plot_test.ylabel('selling_price')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
# Print the model summary statistics. This is where you evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

Since the slope is upward, the connection is also upward.
This supports the hypothesis that a car's price increases with its engine's power.
The regression line is not at all a representation of the data because the MSE is large (520060563679).
R2 value accounts for 18%, which indicates as a poor correlation
